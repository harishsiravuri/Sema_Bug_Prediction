```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("combined_data.csv")
```

```python
summary = data.describe()
summary = summary.transpose()
summary
```

```python
corr_mat = data.corr()
corr_mat
```

```python
indices = np.where(corr_mat > 0.7)
indices = [(corr_mat.index[x], corr_mat.columns[y]) for x, y in zip(*indices) if x != y and x < y]
indices
```

```python
corr_matrix = data.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
to_drop
```

```python
data_no_corr_95 = data
data_no_corr_95.drop(to_drop, axis = 1)
data_no_corr_95
```

```python
data_no_corr_85 = data
data_no_corr_85.drop(to_drop, axis = 1)
data_no_corr_85
```
